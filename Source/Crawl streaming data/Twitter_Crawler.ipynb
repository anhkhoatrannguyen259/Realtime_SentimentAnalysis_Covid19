```python
#override tweepy.StreamListener to add logic to on_status
i=0
class MyStreamListener(tweepy.StreamListener):
    
    def on_status(self, status):  
      global i      
      tweet = status._json
      if not status.truncated:
        text = status.text
      else:
        text = status.extended_tweet['full_text']
      tweet['full_text'] = text 
      if tweet['lang'] =='en' and tweet['full_text'][:2] != 'RT' and tweet['full_text'][:3] != 'Pin' and "stream" not in tweet['full_text'].lower() and "song" not in tweet['full_text'].lower():           
        if "sinopharm" in text.lower():
          tweet["Sinopharm"] = '1'
        else:
          tweet["Sinopharm"] = '0'  

        if "pfizer" in text.lower():
          tweet["Pfizer"] = '1'
        else:
          tweet["Pfizer"] = '0'  

        if "sinovac" in text.lower():
          tweet["Sinovac"] = '1'
        else:
          tweet["Sinovac"] = '0'  

        if "moderna" in text.lower():
          tweet["Moderna"] = '1'
        else:
          tweet["Moderna"] = '0'  

        if "astrazeneca" in text.lower():
          tweet["AstraZeneca"] = '1'
        else:
          tweet["AstraZeneca"] = '0'  

        if "covaxin" in text.lower():
          tweet["Covaxin"] = '1'  
        else:
          tweet["Covaxin"] = '0'  

        if "sputnik" in text.lower():
          tweet["Sputnik"] = '1'
        else:
          tweet["Sputnik"] = '0'              
        with open(f"/content/drive/MyDrive/TwitterData/tw{i}.json", 'w') as f:
          json.dump(tweet, f)
        print(f"{i} - {tweet['created_at']}--{tweet['full_text']}")
        i+=1
```

```python
myStreamListener = MyStreamListener()
myStream = tweepy.Stream(auth = api.auth, listener=myStreamListener, tweet_mode='extended')
```

```python
myStream.filter(track=['Sinopharm','Pfizer','Sinovac','Moderna','AstraZeneca','Covaxin','Sputnik'])
```
